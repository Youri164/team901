```python
import pandas as pd

# 현재 폴더에서 CSV 파일 읽기
df = pd.read_csv(r'C:\Users\dlwlg\team901\learning2_data_cleaned.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0'], inplace=True, axis=1)
```

```python
df.head()
```

```python
signal_data = df.drop(columns=['fault_type'])
```

```python
signal_data
```

```python
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
signal_data_scaled = scaler.fit_transform(signal_data)
```

```python
svm_model = OneClassSVM(nu=0.05, kernel="rbf", gamma='scale')
svm_model.fit(signal_data_scaled)
```

```python
outlier_predictions = svm_model.predict(signal_data_scaled)
```

```python
df['outlier'] = outlier_predictions
```

```python
df['outlier'].value_counts()
```

```python
print(df[['fault_type', 'outlier']])
```
