```python
import pandas as pd
import numpy as np

#     
def calculate_fft_magnitude(data_group, sampling_rate, target_freqs, bandwidth=30):
    fft_magnitudes = []
    frequencies = np.fft.fftfreq(data_group.shape[1], d=1/sampling_rate)
    
    for _, signal in data_group.iterrows():
        # FFT 
        fft_values = np.fft.fft(signal)
        fft_magnitude = np.abs(fft_values)[:len(fft_values)//2]  #   
        pos_frequencies = frequencies[:len(frequencies)//2]  #  

        group_magnitudes = {}
        for freq in target_freqs:
            #   :   0Hz 
            low_bound = max(0, freq - bandwidth)
            high_bound = freq + bandwidth

            #   
            freq_range = (pos_frequencies >= low_bound) & (pos_frequencies <= high_bound)
            #    
            magnitude_in_range = np.sum(fft_magnitude[freq_range])
            group_magnitudes[f'Freq_{freq}Hz'] = magnitude_in_range
        
        fft_magnitudes.append(group_magnitudes)
    
    return pd.DataFrame(fft_magnitudes)

#     
sampling_rate = 12000  # Hz
fault_frequencies = {
    'DE': {
        1797: {'FTF': 13.14, 'BSF': 155.49, 'BPFO': 118.19, 'BPFI': 178.93},
        1772: {'FTF': 11.76, 'BSF': 139.22, 'BPFO': 105.89, 'BPFI': 159.99},
        1750: {'FTF': 11.58, 'BSF': 137.64, 'BPFO': 104.41, 'BPFI': 157.76},
        1730: {'FTF': 11.42, 'BSF': 136.25, 'BPFO': 103.06, 'BPFI': 155.63}
    },
    'FE': {
        1797: {'FTF': 12.58, 'BSF': 131.44, 'BPFO': 100.65, 'BPFI': 163.10},
        1772: {'FTF': 11.25, 'BSF': 117.63, 'BPFO': 90.07, 'BPFI': 146.04},
        1750: {'FTF': 11.10, 'BSF': 115.97, 'BPFO': 88.75, 'BPFI': 144.14},
        1730: {'FTF': 10.96, 'BSF': 114.34, 'BPFO': 87.46, 'BPFI': 142.24}
    }
}

# fault_type, rpm, classification    
dataframes = {}
for (classification, rpm, fault_type), group in cwru.groupby(['classification', 'rpm', 'fault_type']):
    #   
    df_name = f"{fault_type}_{rpm}_{classification}"
    
    #  rpm  classification    
    target_freqs = list(fault_frequencies[classification][rpm].values())
    
    # fault_type 1 inner, 2 roller, 3 outer, 0 normal  
    prefix = {1: "inner", 2: "roller", 3: "outer", 0: "normal"}.get(fault_type, "unknown")
    df_name = f"{prefix}{rpm}"
    
    #     FFT     
    signal_data = group.drop(columns=['rpm', 'classification', 'fault_type'])
    fft_magnitudes = calculate_fft_magnitude(signal_data, sampling_rate, target_freqs)
    
    # fault_type, rpm, classification  
    fft_magnitudes['fault_type'] = fault_type
    fft_magnitudes['rpm'] = rpm
    fft_magnitudes['classification'] = classification
    
    #    
    dataframes[df_name] = fft_magnitudes

#    
final_df = pd.concat(dataframes.values(), ignore_index=True)

#   
final_df.columns = ['ftf_freq', 'bsf_freq', 'bpfo_freq', 'bpfi_freq', 'fault_type', 'rpm', 'classification']

#  
print(final_df.head())
```
