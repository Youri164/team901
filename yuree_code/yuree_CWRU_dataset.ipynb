```python
import pandas as pd
import numpy as np
from scipy.fft import fft, fftfreq

# 샘플링 레이트와 샘플 개수 설정
sampling_rate = 12000  # Hz
num_samples = sampling_rate * 1  # 1초 신호

# DE와 FE별로 rpm에 따른 결함 주파수 설정
fault_frequencies = {
    'DE': {
        1797: {'FTF': 13.14, 'BSF': 155.49, 'BPFO': 118.19, 'BPFI': 178.93},
        1772: {'FTF': 11.76, 'BSF': 139.22, 'BPFO': 105.89, 'BPFI': 159.99},
        1750: {'FTF': 11.58, 'BSF': 137.64, 'BPFO': 104.41, 'BPFI': 157.76},
        1730: {'FTF': 11.42, 'BSF': 136.25, 'BPFO': 103.06, 'BPFI': 155.63}
    },
    'FE': {
        1797: {'FTF': 12.58, 'BSF': 131.44, 'BPFO': 100.65, 'BPFI': 163.10},
        1772: {'FTF': 11.25, 'BSF': 117.63, 'BPFO': 90.07, 'BPFI': 146.04},
        1750: {'FTF': 11.10, 'BSF': 115.97, 'BPFO': 88.75, 'BPFI': 144.14},
        1730: {'FTF': 10.96, 'BSF': 114.34, 'BPFO': 87.46, 'BPFI': 142.24}
    }
}

# 결과 저장을 위한 빈 리스트 생성
results = []

# 각 row에 대해 FFT 수행 및 관심 주파수에서 크기 추출
for idx, row in cwru.iterrows():
    rpm = row['rpm']
    classification = row['classification']
    fault_type = row['fault_type']
    
    # 분류와 rpm에 따라 결함 주파수 선택
    freqs = fault_frequencies[classification][rpm]
    
    # 신호 데이터 추출: '0'부터 '11999' 열까지의 데이터를 하나의 배열로 결합
    signal = row.loc['0':'11999'].values.astype(float)
    
    # FFT 변환
    fft_result = np.abs(fft(signal))
    fft_freqs = fftfreq(num_samples, 1 / sampling_rate)
    
    # 결함 주파수에서 크기 추출
    magnitude_data = {}
    for fault, freq in freqs.items():
        idx = np.argmin(np.abs(fft_freqs - freq))
        magnitude_data[fault] = fft_result[idx]

    # 새로운 행 추가
    magnitude_data['Hz'] = sampling_rate
    magnitude_data['RPM'] = rpm
    magnitude_data['fault_type'] = fault_type
    results.append(magnitude_data)

# 결과 데이터프레임 생성
fault_frequency = pd.DataFrame(results)

# 결과 확인
print(fault_frequency.head())
```

```python
import pandas as pd
import numpy as np

# 주파수와 진폭의 세기를 계산하는 함수
def calculate_fft_magnitude(data_group, sampling_rate, target_freqs, bandwidth=30):
    fft_magnitudes = []
    frequencies = np.fft.fftfreq(data_group.shape[1], d=1/sampling_rate)
    
    for _, signal in data_group.iterrows():
        # FFT 변환
        fft_values = np.fft.fft(signal)
        fft_magnitude = np.abs(fft_values)[:len(fft_values)//2]  # 양수 주파수만 선택
        pos_frequencies = frequencies[:len(frequencies)//2]  # 양수 주파수

        group_magnitudes = {}
        for freq in target_freqs:
            # 주파수 범위 설정: 음수 주파수는 0Hz로 제한
            low_bound = max(0, freq - bandwidth)
            high_bound = freq + bandwidth

            # 주파수 대역 설정
            freq_range = (pos_frequencies >= low_bound) & (pos_frequencies <= high_bound)
            # 대역 내에서 진폭 합산
            magnitude_in_range = np.sum(fft_magnitude[freq_range])
            group_magnitudes[f'Freq_{freq}Hz'] = magnitude_in_range
        
        fft_magnitudes.append(group_magnitudes)
    
    return pd.DataFrame(fft_magnitudes)

# 샘플링 레이트와 결함 주파수 설정
sampling_rate = 12000  # Hz
fault_frequencies = {
    'DE': {
        1797: {'FTF': 13.14, 'BSF': 155.49, 'BPFO': 118.19, 'BPFI': 178.93},
        1772: {'FTF': 11.76, 'BSF': 139.22, 'BPFO': 105.89, 'BPFI': 159.99},
        1750: {'FTF': 11.58, 'BSF': 137.64, 'BPFO': 104.41, 'BPFI': 157.76},
        1730: {'FTF': 11.42, 'BSF': 136.25, 'BPFO': 103.06, 'BPFI': 155.63}
    },
    'FE': {
        1797: {'FTF': 12.58, 'BSF': 131.44, 'BPFO': 100.65, 'BPFI': 163.10},
        1772: {'FTF': 11.25, 'BSF': 117.63, 'BPFO': 90.07, 'BPFI': 146.04},
        1750: {'FTF': 11.10, 'BSF': 115.97, 'BPFO': 88.75, 'BPFI': 144.14},
        1730: {'FTF': 10.96, 'BSF': 114.34, 'BPFO': 87.46, 'BPFI': 142.24}
    }
}

# fault_type, rpm, classification에 따라 데이터를 나누어 처리
dataframes = {}
for (classification, rpm, fault_type), group in cwru.groupby(['classification', 'rpm', 'fault_type']):
    # 데이터프레임 이름 생성
    df_name = f"{fault_type}_{rpm}_{classification}"
    
    # 해당 rpm 및 classification에 맞는 결함 주파수 선택
    target_freqs = list(fault_frequencies[classification][rpm].values())
    
    # fault_type 1은 inner, 2는 roller, 3은 outer, 0은 normal로 이름 지정
    prefix = {1: "inner", 2: "roller", 3: "outer", 0: "normal"}.get(fault_type, "unknown")
    df_name = f"{prefix}{rpm}"
    
    # 신호 데이터 추출 및 FFT 변환 후 진폭 합산 계산
    signal_data = group.drop(columns=['rpm', 'classification', 'fault_type'])
    fft_magnitudes = calculate_fft_magnitude(signal_data, sampling_rate, target_freqs)
    
    # fault_type, rpm, classification 열 추가
    fft_magnitudes['fault_type'] = fault_type
    fft_magnitudes['rpm'] = rpm
    fft_magnitudes['classification'] = classification
    
    # 결과를 데이터프레임 딕셔너리에 저장
    dataframes[df_name] = fft_magnitudes

# 모든 데이터프레임을 하나로 합치기
final_df = pd.concat(dataframes.values(), ignore_index=True)

# 열 이름 정리
final_df.columns = ['ftf_freq', 'bsf_freq', 'bpfo_freq', 'bpfi_freq', 'fault_type', 'rpm', 'classification']

# 결과 확인
print(final_df.head())
```

```python
# 주파수와 진폭의 세기를 계산하는 함수
def calculate_fft_magnitude(data_group, sampling_rate, target_freqs, bandwidth=30):
    fft_magnitudes = []
    frequencies = np.fft.fftfreq(data_group.shape[1], d=1/sampling_rate)
    
    for _, signal in data_group.iterrows():
        # FFT 변환
        fft_values = np.fft.fft(signal)
        fft_magnitude = np.abs(fft_values)[:len(fft_values)//2]  # 양수 주파수만 선택
        pos_frequencies = frequencies[:len(frequencies)//2]  # 양수 주파수

        # 주파수 대역 내 진폭 합산을 저장할 딕셔너리
        group_magnitudes = {
            'FTF_data': 0,
            'BSF_data': 0,
            'BPFO_data': 0,
            'BPFI_data': 0
        }
        
        # 각 결함 주파수의 진폭 합산 계산
        for i, freq in enumerate(target_freqs):
            # 주파수 범위 설정: 음수 주파수는 0Hz로 제한
            low_bound = max(0, freq - bandwidth)
            high_bound = freq + bandwidth

            # 주파수 대역 설정
            freq_range = (pos_frequencies >= low_bound) & (pos_frequencies <= high_bound)
            # 대역 내에서 진폭 합산
            magnitude_in_range = np.sum(fft_magnitude[freq_range])
            
            # 적절한 결함 주파수에 저장
            if i == 0:
                group_magnitudes['FTF_data'] = magnitude_in_range
            elif i == 1:
                group_magnitudes['BSF_data'] = magnitude_in_range
            elif i == 2:
                group_magnitudes['BPFO_data'] = magnitude_in_range
            elif i == 3:
                group_magnitudes['BPFI_data'] = magnitude_in_range
        
        fft_magnitudes.append(group_magnitudes)
    
    return pd.DataFrame(fft_magnitudes)

# 샘플링 레이트와 결함 주파수 설정
sampling_rate = 12000  # Hz
fault_frequencies = {
    'DE': {
        1797: {'FTF': 13.14, 'BSF': 155.49, 'BPFO': 118.19, 'BPFI': 178.93},
        1772: {'FTF': 11.76, 'BSF': 139.22, 'BPFO': 105.89, 'BPFI': 159.99},
        1750: {'FTF': 11.58, 'BSF': 137.64, 'BPFO': 104.41, 'BPFI': 157.76},
        1730: {'FTF': 10.96, 'BSF': 114.34, 'BPFO': 87.46, 'BPFI': 142.24}
    },
    'FE': {
        1797: {'FTF': 12.58, 'BSF': 131.44, 'BPFO': 100.65, 'BPFI': 163.10},
        1772: {'FTF': 11.25, 'BSF': 117.63, 'BPFO': 90.07, 'BPFI': 146.04},
        1750: {'FTF': 11.10, 'BSF': 115.97, 'BPFO': 88.75, 'BPFI': 144.14},
        1730: {'FTF': 10.96, 'BSF': 114.34, 'BPFO': 87.46, 'BPFI': 142.24}
    }
}

# fault_type, rpm, classification에 따라 데이터를 나누어 처리
dataframes = {}
for (classification, rpm, fault_type), group in cwru.groupby(['classification', 'rpm', 'fault_type']):
    # 데이터프레임 이름 생성
    df_name = f"{fault_type}_{rpm}"
    
    # 해당 rpm 및 classification에 맞는 결함 주파수 선택
    target_freqs = list(fault_frequencies[classification][rpm].values())
    
    # fault_type 1은 inner, 2는 roller, 3은 outer, 0은 normal로 이름 지정
    prefix = {1: "inner", 2: "roller", 3: "outer", 0: "normal"}.get(fault_type, "unknown")
    df_name = f"{prefix}{rpm}"
    
    # 신호 데이터 추출 및 FFT 변환 후 진폭 합산 계산
    signal_data = group.drop(columns=['rpm', 'classification', 'fault_type'])
    fft_magnitudes = calculate_fft_magnitude(signal_data, sampling_rate, target_freqs)
    
    # fault_type, rpm, classification 열 추가
    fft_magnitudes['fault_type'] = fault_type
    fft_magnitudes['rpm'] = rpm
    fft_magnitudes['classification'] = classification
    
    # 결과를 데이터프레임 딕셔너리에 저장
    dataframes[df_name] = fft_magnitudes

# 모든 데이터프레임을 하나로 합치기
final_df = pd.concat(dataframes.values(), ignore_index=True)

# 열 이름 확인
print(final_df.head())

```
