```python
import math

def calculate_fault_frequency(rpm, ball_count, contact_angle, ball_diameter, pitch_diameter):
    rotation_frequency = rpm / 60.0
    contact_angle_rad = math.radians(contact_angle)

    # BPFO (Ball Pass Frequency Outer Race)
    bpfo = (rotation_frequency * ball_count) / 2 * (1 - (ball_diameter / pitch_diameter) * math.cos(contact_angle_rad))
    
    # BPFI (Ball Pass Frequency Inner Race)
    bpfi = (rotation_frequency * ball_count) / 2 * (1 + (ball_diameter / pitch_diameter) * math.cos(contact_angle_rad))
    
    # BSF (Ball Spin Frequency)
    bsf = (pitch_diameter * rotation_frequency) / (2 * ball_diameter) * (1 - ((ball_diameter / pitch_diameter) * math.cos(contact_angle_rad))**2)
    
    # FTF (Fundamental Train Frequency)
    ftf = rotation_frequency / 2 * (1 - (ball_diameter / pitch_diameter) * math.cos(contact_angle_rad))

    print(f"Calculated BPFO: {bpfo:.2f} Hz, BPFI: {bpfi:.2f} Hz, BSF: {bsf:.2f} Hz, FTF: {ftf:.2f} Hz")

    return bpfo, bpfi, bsf, ftf


# --------- 주파수에 따른 진폭 함수 ---------
def calculate_fault_frequency_magnitude(data_group, fs, rpm, ball_count, contact_angle, ball_diameter, pitch_diameter, bandwidth=30):
    # Calculate BPFO, BPFI, BSF, and FTF frequencies
    bpfo_freq, bpfi_freq, bsf_freq, ftf_freq = calculate_fault_frequency(rpm, ball_count, contact_angle, ball_diameter, pitch_diameter)
    
    fft_magnitudes = []
    frequencies = np.fft.fftfreq(data_group.shape[1] - 1, d=1/fs)
    
    for index, signal in data_group.iloc[:, :-1].iterrows():
        fft_values = np.fft.fft(signal)
        fft_magnitude = np.abs(fft_values)[:len(fft_values)//2]
        pos_frequencies = frequencies[:len(frequencies)//2]

        group_magnitudes = {}
        
        for freq, label in zip([bpfo_freq, bpfi_freq, bsf_freq, ftf_freq], ["BPFO_data", "BPFI_data", "BSF_data", "FTF_data"]):
            low_bound = max(0, freq - bandwidth)
            high_bound = freq + bandwidth
            freq_range = (pos_frequencies >= low_bound) & (pos_frequencies <= high_bound)
            magnitude_in_range = np.sum(fft_magnitude[freq_range])
            
            group_magnitudes[label] = magnitude_in_range
        
        fft_magnitudes.append(group_magnitudes)
    
    result_df = pd.DataFrame(fft_magnitudes)
    return result_df
```

```python
import pandas as pd
import numpy as np

# 샘플링 레이트와 결함 주파수 설정
sampling_rate = 48000  # Hz
fault_frequencies = {
    'DE': {
        1797: {'FTF': 13.14, 'BSF': 155.49, 'BPFO': 118.19, 'BPFI': 178.93},
        1772: {'FTF': 11.76, 'BSF': 139.22, 'BPFO': 105.89, 'BPFI': 159.99},
        1750: {'FTF': 11.58, 'BSF': 137.64, 'BPFO': 104.41, 'BPFI': 157.76},
        1730: {'FTF': 10.96, 'BSF': 114.34, 'BPFO': 87.46, 'BPFI': 142.24}
    },
    'FE': {
        1797: {'FTF': 12.58, 'BSF': 131.44, 'BPFO': 100.65, 'BPFI': 163.10},
        1772: {'FTF': 11.25, 'BSF': 117.63, 'BPFO': 90.07, 'BPFI': 146.04},
        1750: {'FTF': 11.10, 'BSF': 115.97, 'BPFO': 88.75, 'BPFI': 144.14},
        1730: {'FTF': 10.96, 'BSF': 114.34, 'BPFO': 87.46, 'BPFI': 142.24}
    }
}

# 주파수와 진폭의 세기를 계산하는 함수
def calculate_fft_magnitude(data_group, sampling_rate, target_freqs, bandwidth=30):
    fft_magnitudes = []
    frequencies = np.fft.fftfreq(data_group.shape[1], d=1/sampling_rate)
    
    for _, signal in data_group.iterrows():
        # FFT 변환
        fft_values = np.fft.fft(signal)
        fft_magnitude = np.abs(fft_values)[:len(fft_values)//2]  # 양수 주파수만 선택
        pos_frequencies = frequencies[:len(frequencies)//2]  # 양수 주파수

        # 주파수 대역 내 진폭 합산을 저장할 딕셔너리
        group_magnitudes = {
            'FTF_data': 0,
            'BSF_data': 0,
            'BPFO_data': 0,
            'BPFI_data': 0
        }
        
        # 각 결함 주파수의 진폭 합산 계산
        for i, freq in enumerate(target_freqs):
            # 주파수 범위 설정: 음수 주파수는 0Hz로 제한
            low_bound = max(0, freq - bandwidth)
            high_bound = freq + bandwidth

            # 주파수 대역 설정
            freq_range = (pos_frequencies >= low_bound) & (pos_frequencies <= high_bound)
            # 대역 내에서 진폭 합산
            magnitude_in_range = np.sum(fft_magnitude[freq_range])
            
            # 적절한 결함 주파수에 저장
            if i == 0:
                group_magnitudes['FTF_data'] = magnitude_in_range
            elif i == 1:
                group_magnitudes['BSF_data'] = magnitude_in_range
            elif i == 2:
                group_magnitudes['BPFO_data'] = magnitude_in_range
            elif i == 3:
                group_magnitudes['BPFI_data'] = magnitude_in_range
        
        fft_magnitudes.append(group_magnitudes)
    
    return pd.DataFrame(fft_magnitudes)

# 원래 데이터에 인덱스 열 추가
cwru48 = cwru48.reset_index().rename(columns={'index': 'original_index'})

# fault_type, rpm, classification에 따라 데이터를 나누어 처리
dataframes = []
for (classification, rpm, fault_type), group in cwru48.groupby(['classification', 'rpm', 'fault_type']):
    # 해당 rpm 및 classification에 맞는 결함 주파수 선택
    target_freqs = list(fault_frequencies[classification][rpm].values())
    
    # 신호 데이터 추출 및 FFT 변환 후 진폭 합산 계산
    signal_data = group.drop(columns=['rpm', 'classification', 'fault_type', 'original_index'])
    fft_magnitudes = calculate_fft_magnitude(signal_data, sampling_rate, target_freqs)
    
    # fault_type, rpm, classification, 원래 인덱스 열 추가 (리스트 형태로 추가하여 1D 배열 유지)
    fft_magnitudes['fault_type'] = fault_type
    fft_magnitudes['rpm'] = rpm
    fft_magnitudes['classification'] = classification
    fft_magnitudes['original_index'] = group['original_index'].values.tolist()  # 1D 배열로 변환하여 추가
    
    # 결과를 데이터프레임 리스트에 추가
    dataframes.append(fft_magnitudes)

# 모든 데이터프레임을 하나로 합치고, 원래 순서대로 정렬
cwru_df = pd.concat(dataframes, ignore_index=True)
cwru_df = cwru_df.sort_values(by='original_index').reset_index(drop=True)

# 필요 없는 'original_index' 열 삭제
cwru_df = cwru_df.drop(columns=['original_index'])
cwru_df = cwru_df.drop(columns=['classification'])
```

```python
import pandas as pd
import numpy as np

# 샘플링 레이트와 결함 주파수 설정
sampling_rate = 48000  # Hz
fault_frequencies = {
    'DE': {
        1797: {'FTF': 13.14, 'BSF': 155.49, 'BPFO': 118.19, 'BPFI': 178.93},
        1772: {'FTF': 11.76, 'BSF': 139.22, 'BPFO': 105.89, 'BPFI': 159.99},
        1750: {'FTF': 11.58, 'BSF': 137.64, 'BPFO': 104.41, 'BPFI': 157.76},
        1730: {'FTF': 10.96, 'BSF': 114.34, 'BPFO': 87.46, 'BPFI': 142.24}
    },
    'FE': {
        1797: {'FTF': 12.58, 'BSF': 131.44, 'BPFO': 100.65, 'BPFI': 163.10},
        1772: {'FTF': 11.25, 'BSF': 117.63, 'BPFO': 90.07, 'BPFI': 146.04},
        1750: {'FTF': 11.10, 'BSF': 115.97, 'BPFO': 88.75, 'BPFI': 144.14},
        1730: {'FTF': 10.96, 'BSF': 114.34, 'BPFO': 87.46, 'BPFI': 142.24}
    }
}

# 주파수와 진폭의 세기를 계산하는 함수
def calculate_fft_magnitude(data_group, sampling_rate, target_freqs, bandwidth=30):
    fft_magnitudes = []
    frequencies = np.fft.fftfreq(data_group.shape[1], d=1/sampling_rate)
    
    for _, signal in data_group.iterrows():
        # FFT 변환
        fft_values = np.fft.fft(signal)
        fft_magnitude = np.abs(fft_values)[:len(fft_values)//2]  # 양수 주파수만 선택
        pos_frequencies = frequencies[:len(frequencies)//2]  # 양수 주파수

        # 주파수 대역 내 진폭 합산을 저장할 딕셔너리
        group_magnitudes = {
            'FTF_data': 0,
            'BSF_data': 0,
            'BPFO_data': 0,
            'BPFI_data': 0
        }
        
        # 각 결함 주파수의 진폭 합산 계산
        for i, freq in enumerate(target_freqs):
            # 주파수 범위 설정: 음수 주파수는 0Hz로 제한
            low_bound = max(0, freq - bandwidth)
            high_bound = freq + bandwidth

            # 주파수 대역 설정
            freq_range = (pos_frequencies >= low_bound) & (pos_frequencies <= high_bound)
            # 대역 내에서 진폭 합산
            magnitude_in_range = np.sum(fft_magnitude[freq_range])
            
            # 적절한 결함 주파수에 저장
            if i == 0:
                group_magnitudes['FTF_data'] = magnitude_in_range
            elif i == 1:
                group_magnitudes['BSF_data'] = magnitude_in_range
            elif i == 2:
                group_magnitudes['BPFO_data'] = magnitude_in_range
            elif i == 3:
                group_magnitudes['BPFI_data'] = magnitude_in_range
        
        fft_magnitudes.append(group_magnitudes)
    
    return pd.DataFrame(fft_magnitudes)

# 원래 데이터에 인덱스 열 추가
normal48 = normal48.reset_index().rename(columns={'index': 'original_index'})

# fault_type, rpm, classification에 따라 데이터를 나누어 처리
dataframes = []
for (classification, rpm, fault_type), group in normal48.groupby(['classification', 'rpm', 'fault_type']):
    # 해당 rpm 및 classification에 맞는 결함 주파수 선택
    target_freqs = list(fault_frequencies[classification][rpm].values())
    
    # 신호 데이터 추출 및 FFT 변환 후 진폭 합산 계산
    signal_data = group.drop(columns=['rpm', 'classification', 'fault_type', 'original_index'])
    fft_magnitudes = calculate_fft_magnitude(signal_data, sampling_rate, target_freqs)
    
    # fault_type, rpm, classification, 원래 인덱스 열 추가 (리스트 형태로 추가하여 1D 배열 유지)
    fft_magnitudes['fault_type'] = fault_type
    fft_magnitudes['rpm'] = rpm
    fft_magnitudes['classification'] = classification
    fft_magnitudes['original_index'] = group['original_index'].values.tolist()  # 1D 배열로 변환하여 추가
    
    # 결과를 데이터프레임 리스트에 추가
    dataframes.append(fft_magnitudes)

# 모든 데이터프레임을 하나로 합치고, 원래 순서대로 정렬
normal_df = pd.concat(dataframes, ignore_index=True)
normal_df = normal_df.sort_values(by='original_index').reset_index(drop=True)

# 필요 없는 'original_index' 열 삭제
normal_df = normal_df.drop(columns=['original_index'])
normal_df = normal_df.drop(columns=['classification'])
```
