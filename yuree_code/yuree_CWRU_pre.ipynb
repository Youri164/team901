## inner

```python
import scipy.io as sio
inner_data = sio.loadmat(r"C:\Users\medici\901\CWRU\DE\1\1730\inner1_1730.mat")
```

```python
inner_data
```

```python
inner_data['X172_DE_time'].shape
```

```python
import pandas as pd

# inner_data의 'X107_DE_time' 열을 1차원으로 변환
inner_data_1d = inner_data['X172_DE_time'].squeeze()

# DataFrame으로 변환
inner_data_df = pd.DataFrame(inner_data_1d, columns=['X172_DE_time'])

# 데이터를 12000개씩 나누어 변수에 저장
inner_data1 = inner_data_df['X172_DE_time'][:12000]
inner_data2 = inner_data_df['X172_DE_time'][12000:24000]
inner_data3 = inner_data_df['X172_DE_time'][24000:36000]
inner_data4 = inner_data_df['X172_DE_time'][36000:48000]
inner_data5 = inner_data_df['X172_DE_time'][48000:60000]
inner_data6 = inner_data_df['X172_DE_time'][60000:72000]
inner_data7 = inner_data_df['X172_DE_time'][72000:84000]
inner_data8 = inner_data_df['X172_DE_time'][84000:96000]
inner_data9 = inner_data_df['X172_DE_time'][96000:108000]
inner_data10 = inner_data_df['X172_DE_time'][108000:120000]
```

```python
inner_data1.shape
```

```python
import pandas as pd
import numpy as np

# List of inner_data arrays (inner_data1 to inner_data10)
inner_data_list = [inner_data1, inner_data2, inner_data3, inner_data4, inner_data5,
                   inner_data6, inner_data7, inner_data8, inner_data9, inner_data10]

# List to store each DataFrame row
dataframes = []

# Loop over each inner_data and create a DataFrame row
for inner_data in inner_data_list:
    # Append rpm and fault_type to each array
    data_array = np.append(inner_data, [1730, 1])
    # Create a DataFrame row and add classification column
    df_row = pd.DataFrame([data_array[:-2]], columns=[f'{j}' for j in range(len(data_array)-2)])
    df_row['rpm'] = 1730
    df_row['fault_type'] = 1
    df_row['classification'] = 'DE'
    dataframes.append(df_row)

# Concatenate all rows into a single DataFrame
df_all = pd.concat(dataframes, ignore_index=True)

# Save the final DataFrame to a CSV file
file_path = r'C:\Users\medici\901\CWRU\inner_1730_2_DE.csv'
df_all.to_csv(file_path, index=False)
```
