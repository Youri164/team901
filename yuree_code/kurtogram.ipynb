```python
import numpy as np
import pandas as pd
from scipy.signal import stft

# 스펙트럼 커터시스의 근사치를 계산하는 함수
def spectral_kurtosis(Zxx):
    magnitude = np.abs(Zxx)
    mean_mag = np.mean(magnitude, axis=1)
    std_mag = np.std(magnitude, axis=1)
    sk = (np.mean((magnitude - mean_mag[:, None])**4, axis=1) / std_mag**4) - 3
    return sk

# 각 행에 대해 상위 10개의 주파수를 열 형식으로 추출하는 함수
def extract_top_frequencies_single(data, fs, nperseg=128):
    top_frequencies_df = pd.DataFrame()  # 결과를 저장할 데이터프레임

    for index, row in data.iterrows():
        # 각 행 데이터에 대해 STFT와 스펙트럼 커터시스 계산
        f, t, Zxx = stft(row.values, fs=fs, nperseg=nperseg)
        sk_values = spectral_kurtosis(Zxx)

        # 상위 10개의 커터시스 값에 해당하는 주파수 추출
        top_10_indices = np.argsort(sk_values)[-10:][::-1]  # 상위 10개 인덱스 (내림차순 정렬)
        top_10_frequencies = f[top_10_indices]

        # 주파수를 열 형식으로 저장
        top_frequencies = {f'frequency{i+1}': top_10_frequencies[i] for i in range(10)}
        top_frequencies_df = pd.concat([top_frequencies_df, pd.DataFrame([top_frequencies])], ignore_index=True)
    
    return top_frequencies_df

# 샘플 데이터 준비
fs = 48828  # 샘플링 주파수
data488 = mfpt488.iloc[:42, :]  # 480개의 행 전체 선택

# 함수 실행 및 결과 저장
top_frequencies_488 = extract_top_frequencies_single(data488, fs)

# 결과 출력
top_frequencies_488
```

```python
import numpy as np
import pandas as pd
from scipy.signal import stft

# 스펙트럼 커터시스의 근사치를 계산하는 함수
def spectral_kurtosis(Zxx):
    magnitude = np.abs(Zxx)
    mean_mag = np.mean(magnitude, axis=1)
    std_mag = np.std(magnitude, axis=1)
    sk = (np.mean((magnitude - mean_mag[:, None])**4, axis=1) / std_mag**4) - 3
    return sk

# 각 행에 대해 전체 데이터를 사용하여 중심 주파수를 계산하는 함수
def calculate_center_frequencies(data, fs, nperseg=128):
    center_frequencies = []

    for index, row in data.iterrows():
        f, t, Zxx = stft(row[:-1], fs=fs, nperseg=nperseg)  # 'fault_type' 제외한 데이터 사용
        sk_values = spectral_kurtosis(Zxx)
        
        # 중심 주파수 계산
        center_freq = f[np.argmax(sk_values)]
        center_frequencies.append(center_freq)

    return center_frequencies

# 샘플링 주파수 설정
fs_data = 25600
fs_cwru = 12000

# 각 데이터셋에서 fault_type 열을 제외하고 중심 주파수를 계산하여 데이터프레임으로 저장
data_center = pd.DataFrame({'Center Frequency': calculate_center_frequencies(data.iloc[:, :-1], fs=fs_data)})
data2_center = pd.DataFrame({'Center Frequency': calculate_center_frequencies(data2.iloc[:, :-1], fs=fs_data)})
cwru_center = pd.DataFrame({'Center Frequency': calculate_center_frequencies(cwru.iloc[:, :-1], fs=fs_cwru)})
```
