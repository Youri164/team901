```python
import numpy as np
import pandas as pd
from scipy.signal import stft

# ìŠ¤íŽ™íŠ¸ëŸ¼ ì»¤í„°ì‹œìŠ¤ì˜ ê·¼ì‚¬ì¹˜ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜
def spectral_kurtosis(Zxx):
    magnitude = np.abs(Zxx)
    mean_mag = np.mean(magnitude, axis=1)
    std_mag = np.std(magnitude, axis=1)
    sk = (np.mean((magnitude - mean_mag[:, None])**4, axis=1) / std_mag**4) - 3
    return sk

# ê° í–‰ì— ëŒ€í•´ ì „ì²´ ë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ ì¤‘ì‹¬ ì£¼íŒŒìˆ˜ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜
def calculate_center_frequencies(data, fs, nperseg=128):
    center_frequencies = []

    for index, row in data.iterrows():
        f, t, Zxx = stft(row[:-1], fs=fs, nperseg=nperseg)  # 'fault_type' ì œì™¸í•œ ë°ì´í„° ì‚¬ìš©
        sk_values = spectral_kurtosis(Zxx)
        
        # ì¤‘ì‹¬ ì£¼íŒŒìˆ˜ ê³„ì‚°
        center_freq = f[np.argmax(sk_values)]
        center_frequencies.append(center_freq)

    return center_frequencies

# ìƒ˜í”Œë§ ì£¼íŒŒìˆ˜ ì„¤ì •
fs_data = 25600
fs_cwru = 12000

# ê° ë°ì´í„°ì…‹ì—ì„œ fault_type ì—´ì„ ì œì™¸í•˜ê³  ì¤‘ì‹¬ ì£¼íŒŒìˆ˜ë¥¼ ê³„ì‚°í•˜ì—¬ ë°ì´í„°í”„ë ˆìž„ìœ¼ë¡œ ì €ìž¥
data_center = pd.DataFrame({'Center Frequency': calculate_center_frequencies(data.iloc[:, :-1], fs=fs_data)})
data2_center = pd.DataFrame({'Center Frequency': calculate_center_frequencies(data2.iloc[:, :-1], fs=fs_data)})
cwru_center = pd.DataFrame({'Center Frequency': calculate_center_frequencies(cwru.iloc[:, :-1], fs=fs_cwru)})
```
