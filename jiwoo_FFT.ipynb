```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("learning1_data.csv")
```

```python
# 각 열의 널(결측) 값 수 확인
data.isnull().sum().sum()
```

## target = 1(내륜)

```python
## 내륜 1, 롤러 2, 외륜 3, 정상 0
```

```python
# Extract the first row (excluding the target column)
first_row = data.iloc[0, :-1].values
```

```python
# Plot the raw sensor data of the first row
plt.figure(figsize=(10, 6))
plt.plot(first_row)
plt.title('Raw Sensor Data of First Row')
plt.xlabel('Time (Sample Points)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Perform FFT on the data
fft_values = np.fft.fft(first_row)
fft_magnitude = np.abs(fft_values)  # Get magnitude of the FFT
n = len(first_row)
frequencies = np.fft.fftfreq(n)  # Get frequency bins

# Plot the frequency spectrum
plt.figure(figsize=(10, 6))
plt.plot(frequencies[:n // 2], fft_magnitude[:n // 2])  # Plot only positive frequencies
plt.title('Frequency Spectrum of First Row')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
# Sampling frequency
sampling_rate = 25600  # Hz

# Perform FFT on the data
fft_values = np.fft.fft(first_row)
fft_magnitude = np.abs(fft_values)  # Get magnitude of the FFT
n = len(first_row)
frequencies = np.fft.fftfreq(n, d=1/sampling_rate)  # Get frequency bins with sampling rate

# Plot the frequency spectrum
plt.figure(figsize=(10, 6))
plt.plot(frequencies[:n // 2], fft_magnitude[:n // 2])  # Plot only positive frequencies
plt.title('Frequency Spectrum of First Row')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

## 비교

```python
#target = 1
# 행 번호 설정
rows_to_plot = [1, 10, 50]

# 서브플롯 초기화
fig, axs = plt.subplots(len(rows_to_plot), 1, figsize=(10, 12))

# 각 행에 대해 FFT 수행 및 플로팅
for i, row_num in enumerate(rows_to_plot):
    row_data = data.iloc[row_num - 1, :-1].values  # 0-indexed, so subtract 1
    fft_values = np.fft.fft(row_data)
    fft_magnitude = np.abs(fft_values)
    n = len(row_data)
    frequencies = np.fft.fftfreq(n)  # Get frequency bins

    # 서브플롯에 주파수 스펙트럼 추가
    axs[i].plot(frequencies[:n // 2], fft_magnitude[:n // 2])
    axs[i].set_title(f'Frequency Spectrum of Row {row_num}')
    axs[i].set_xlabel('Frequency (Hz)')
    axs[i].set_ylabel('Magnitude')
    axs[i].grid(True)

# 서브플롯 간격 조정
plt.tight_layout()
plt.show()
```

```python
# 샘플링 주파수 설정
sampling_rate = 25600  # Hz

# 행 번호 설정
rows_to_plot = [10, 130, 250, 370]

# 서브플롯 초기화
fig, axs = plt.subplots(len(rows_to_plot), 1, figsize=(10, 12))

# 각 행에 대해 FFT 수행 및 플로팅
for i, row_num in enumerate(rows_to_plot):
    row_data = data.iloc[row_num - 1, :-1].values  # 0-indexed, so subtract 1
    fft_values = np.fft.fft(row_data)
    fft_magnitude = np.abs(fft_values)
    n = len(row_data)
    
    # 샘플링 주파수에 따라 주파수 벡터 계산
    frequencies = np.fft.fftfreq(n, d=1/sampling_rate)  # Get frequency bins with sampling rate

    # 서브플롯에 주파수 스펙트럼 추가
    axs[i].plot(frequencies[:n // 2], fft_magnitude[:n // 2])
    axs[i].set_title(f'Frequency Spectrum of Row {row_num}')
    axs[i].set_xlabel('Frequency (Hz)')
    axs[i].set_ylabel('Magnitude')
    axs[i].grid(True)

# 서브플롯 간격 조정
plt.tight_layout()
plt.show()
```
